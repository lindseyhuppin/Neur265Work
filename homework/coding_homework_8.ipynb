```python
#2 importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
#3 importing data
fear_traces = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/fear_conditioning_traces_hw8.csv', delimiter = ',')
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/extinction_predictors_hw8.csv')
shock_index = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/tone_index.csv', delimiter = ',').astype(int)
extinction_predictors.shape
```

```python
#5 organizing data into conditioning and extinction
conditioning_traces = fear_traces[:10458]
extinction_traces = fear_traces[10458:]
```

```python
#6 printing sizes of arrays
print(conditioning_traces.size)
print(extinction_traces.size)
```

```python
#7 creating time variables
time_conditioning = np.linspace(0,conditioning_traces.shape[0]/15,conditioning_traces.shape[0])
time_extinction = np.linspace(0,extinction_traces.shape[0]/15,extinction_traces.shape[0])
```

The number of ROIs imaged in the session was 63 cells. The conditioning session lasted 697.133 seconds or 11.62 mins. The extinction session was 947.267 seconds or 15.78 mins.

```python
#9 plotting the ROIs
plt.plot(time_conditioning,conditioning_traces,'#0000ff5d')
plt.vlines(shock_index, -.2, 1.2, 'k', lw=2)
plt.vlines(tone_index, -.2, 1.2, 'g', lw=2)
plt.xlabel('time(s)')
plt.ylabel('traces')
plt.title('First ROI')
plt.show()
plt.plot(time_extinction,extinction_traces,'#ff00005d')
plt.vlines(shock_index, -.2, 1.2, 'k', lw=2)
plt.vlines(tone_index, -.2, 1.2, 'g', lw=2)
plt.xlabel('time(s)')
plt.ylabel('traces')
plt.title('Second ROI')
plt.show()
```

###10
During conditioning, the cells are activated after the shock and not necessarily after the tone. However, in the extinction, the cells are activated after the tone instead of only after the shock like in the first region. During conditioning there is no response to the tone, but during extintion the response happens to the tone now that the association is made.

```python
#12 comparing the prediction to the data
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces[:,0]

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})

predictors_conditioning['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
model3_results = model3.fit() # Fit model to our data
b1 = model3_results.params
print(b1)
```

```python
#13 getting correlation coefficient from model
predictors_conditioning['shock onset'] = shock_array
r_index_conditioning = np.zeros(conditioning_traces.shape[1])

for i in range(conditioning_traces.shape[1]):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_index_conditioning[i] = r3.statistic
```

```python
#14 find the correlation coefficient of the extinction data
shock_array = np.zeros(extinction_traces.shape[0])
for i in range(extinction_traces.shape[0]):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(extinction_traces.shape[0])
for i in range(extinction_traces.shape[0]):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0
from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF
first_trace = extinction_traces[:,0]

predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})

predictors_extinction['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors_extinction, family=Poisson())
model3_results = model3.fit() # Fit model to our data
b1 = model3_results.params

predictors_extinction['shock onset'] = shock_array
r_index_extinction = np.zeros(extinction_traces.shape[1])

for i in range(extinction_traces.shape[1]):
  trace_temp = extinction_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_index_extinction[i] = r3.statistic
```

```python
#15 correlate the extinction and conditioning correlation coefficients
print(pearsonr(r_index_conditioning,r_index_extinction))
```

```python
#16 plotting the correlation coefficients
plt.scatter(r_index_conditioning,r_index_extinction)
plt.xlabel('r_index_conditioning')
plt.ylabel('r_index_extinction')
```

According to the plot and the correlation coefficients, there is a correlation between the conditioning and the extinction because the p value is less than .05(.000307<.05).
