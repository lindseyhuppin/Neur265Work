```python
#2 importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3 importing data
spikes_left_trial1 = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/spikes_left_trial1.csv',delimiter = ',')
spikes_left_trial2 = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/spikes_left_trial2.csv',delimiter = ',')
spikes_right_trial1 = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/spikes_right_trial1.csv',delimiter = ',')
spikes_right_trial2 = np.loadtxt('https://raw.githubusercontent.com/lindseyhuppin/Neur265Work/main/spikes_right_trial2.csv',delimiter = ',')
```

```python
#4 printing lengths
print(spikes_left_trial1.size)
print(spikes_left_trial2.size)
print(spikes_right_trial1.size)
print(spikes_right_trial2.size)
```

```python
#6 finding firing rate
firingRate_left1 = spikes_left_trial1.size/30
firingRate_right1 = spikes_right_trial1.size/30
firingRate_left2 = spikes_left_trial2.size/30
firingRate_right2 = spikes_right_trial2.size/30
print(firingRate_left1)
print(firingRate_right1)
print(firingRate_left2)
print(firingRate_right2)
```

```python
#8 plotting spike train data
plt.plot(spikes_left_trial1, np.ones_like(spikes_left_trial1),'y.')
plt.plot(spikes_right_trial1, 1.5*np.ones_like(spikes_right_trial1),'b.')
plt.xlabel('Time(s)')
plt.ylim(0,2.5)
plt.legend(['Left','Right'])
plt.show()
plt.plot(spikes_left_trial2, np.ones_like(spikes_left_trial2),'y.')
plt.plot(spikes_right_trial2, 1.5*np.ones_like(spikes_right_trial2),'b.')
plt.xlabel('Time(s)')
plt.ylim(0,2.5)
plt.legend(['Left','Right'])
plt.show()
```

####9
The left trials seem more bursty than the right ones. The right ones have some bursts but are generally more consistent than the left ones. The left ones have more quiet time than the right ones.

```python
#10 calculate ISI
ISI_left1 = np.diff(spikes_left_trial1)
ISI_right1 = np.diff(spikes_right_trial1)
ISI_left2 = np.diff(spikes_left_trial2)
ISI_right2 = np.diff(spikes_right_trial2)
```

```python
#11 plotting histograms with bins
bins = np.arange(0, 2, .01)
fig,ax = plt.subplots(2,2,figsize = (10,9))
ax[0,0].hist(ISI_left1,bins)
ax[0,1].hist(ISI_right1,bins)
ax[1,0].hist(ISI_left2,bins)
ax[1,1].hist(ISI_right2,bins)
ax[0,0].set_xlabel('ISI(s)')
ax[0,0].set_ylabel('Count')
ax[1,0].set_xlabel('ISI(s)')
ax[1,0].set_ylabel('Count')
ax[0,1].set_xlabel('ISI(s)')
ax[0,1].set_ylabel('Count')
ax[1,1].set_xlabel('ISI(s)')
ax[1,1].set_ylabel('Count')
ax[0,0].set_title('Left Trial 1')
ax[1,0].set_title('Left Trial 2')
ax[0,1].set_title('Right Trial 1')
ax[1,1].set_title('Right Trial 2')
```

####12
Both the right and the left seem to have a lot of burstiness. However, the left trials have more burstiness since it's more concentrated closer to zero. The right ones are pretty concentrated around zero but also have more distribution at higher ISI values, which shows less bursts likely occured for the rights than the lefts.

```python
#13 calculate firing rates
binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_left = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_indexleft = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_valuesleft = spikes_left_trial1[spk_indexleft]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_left[i] = (len(true_valuesleft))/binsize                                      # Divide number of spikes (timestamps) by our bin size to get firing rate

fr_right = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_indexright = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_valuesright = spikes_right_trial1[spk_indexright]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_right[i] = (len(true_valuesright))/binsize                                      # Divide number of spikes (timestamps) by our bin size to get firing rate

plt.plot(fr_bins[:-1], fr_left,'b')
plt.plot(fr_bins[:-1], fr_right,'r')
plt.xlabel('Time [s]')
plt.ylabel('Firing Rate [Hz]')
plt.title('Trial 1')
plt.show()

fr_left = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_indexleft = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_valuesleft = spikes_left_trial2[spk_indexleft]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_left[i] = (len(true_valuesleft))/binsize                                      # Divide number of spikes (timestamps) by our bin size to get firing rate

fr_right = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_indexright = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_valuesright = spikes_right_trial2[spk_indexright]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_right[i] = (len(true_valuesright))/binsize                                      # Divide number of spikes (timestamps) by our bin size to get firing rate

plt.plot(fr_bins[:-1], fr_left,'b')
plt.plot(fr_bins[:-1], fr_right,'r')
plt.xlabel('Time [s]')
plt.ylabel('Firing Rate [Hz]')
plt.title('Trial 2')
plt.show()
```

####14
Generally when the left trial has higher firing rates, the right ones has low firing rates and vice versa. Basically, their peaks are at different points between the right and the left trials. The firing rates also varies a lot for the left turns between trials since trial spikes the most in the first few seconds and around 10-15 seconds while the trial 2 spikes the most after 20 seconds. The right turns both had the highest firing rates before 15 seconds, so there is a lot more similarity between the right turns than the left turns.
